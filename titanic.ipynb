```python
##################################################
# Pandas Altrmalar
##################################################

import numpy as np
import seaborn as sns
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

#########################################
# Grev 1: Seaborn ktphanesi ierisinden Titanic veri setini tanmlaynz.
#########################################
titanic = sns.load_dataset('titanic')
print(titanic.head())

#########################################
# Grev 2: Yukarda tanmlanan Titanic veri setindeki kadn ve erkek yolcularn saysn bulunuz.
#########################################

gender_counts = titanic["sex"].value_counts()
print(gender_counts)


#########################################
# Grev 3: Her bir sutuna ait unique deerlerin saysn bulunuz.
#########################################
titanic.nunique()


#########################################
# Grev 4: pclass deikeninin unique deerleri bulunuz.
#########################################

titanic["pclass"].unique()


#########################################
# Grev 5:  pclass ve parch deikenlerinin unique deerlerinin saysn bulunuz.
#########################################

titanic[["pclass","parch"]].nunique()

#########################################
# Grev 6: embarked deikeninin tipini kontrol ediniz. Tipini category olarak deitiriniz. Tekrar tipini kontrol ediniz.
#########################################

print(titanic["embarked"].dtype)
titanic["embarked"] = titanic["embarked"].astype("category")
print(titanic["embarked"].dtype)
#########################################
# Grev 7: embarked deeri C olanlarn tm bilgelerini gsteriniz.
#########################################

print(titanic[titanic["embarked"]=="C"])


#########################################
# Grev 8: embarked deeri S olmayanlarn tm bilgelerini gsteriniz.
#########################################

print(titanic[titanic["embarked"]!="S"])

#########################################
# Grev 9: Ya 30 dan kk ve kadn olan yolcularn tm bilgilerini gsteriniz.
#########################################

titanic[(titanic["sex"]=="female") & (titanic["age"]<30)]

#########################################
# Grev 10: Fare'i 500'den byk veya ya 70 den byk yolcularn bilgilerini gsteriniz.
#########################################

titanic[(titanic["fare"]>500) | (titanic["age"] > 70)]

#########################################
# Grev 11: Her bir deikendeki bo deerlerin toplamn bulunuz.
#########################################

titanic.isnull().sum()


#########################################
# Grev 12: who deikenini dataframe'den drn.
#########################################

titanic = titanic.drop(columns=['who'])
print(titanic)

#########################################
# Grev 13: deck deikenindeki bo deerleri deck deikenin en ok tekrar eden deeri (mode) ile doldurunuz.
#########################################

deck_mode = titanic["deck"].mode()[0]
titanic["deck"] = titanic["deck"].fillna(deck_mode)
print(titanic)



#########################################
# Grev 14: age deikenindeki bo deerleri age deikenin medyan ile doldurun.
#########################################
age_median = titanic["age"].median()
print(age_median)
titanic["age"] = titanic["age"].fillna(age_median)
print(titanic["age"])


#########################################
# Grev 15: survived deikeninin Pclass ve Cinsiyet deikenleri krlmnnda sum, count, mean deerlerini bulunuz.
#########################################
grouped_data = titanic.groupby(['pclass', 'sex']).agg({
    'survived': ['sum', 'count', 'mean']
})
print(grouped_data)
#########################################
# Grev 16:  30 yan altnda olanlar 1, 30'a eit ve stnde olanlara 0 vericek bir fonksiyon yaznz.
# Yazdnz fonksiyonu kullanarak titanik veri setinde age_flag adnda bir deiken oluturunuz oluturunuz. (apply ve lambda yaplarn kullannz)
#########################################
def age_categorize(age):
    if age < 30:
        return 1
    else:
        return 0
titanic["age_flag"] = titanic["age"].apply(lambda x : age_categorize(x))
print(titanic["age_flag"])

#########################################
# Grev 17: Seaborn ktphanesi ierisinden Tips veri setini tanmlaynz.
#########################################

import seaborn as sns
tips = sns.load_dataset('tips')
print(tips.head())


#########################################
# Grev 18: Time deikeninin kategorilerine (Dinner, Lunch) gre total_bill  deerlerinin toplamn, min, max ve ortalamasn bulunuz.
#########################################

time_group = tips.groupby("time")["total_bill"].agg(["sum", "min", "max", "mean"])
print(time_group)

#########################################
# Grev 19: Gnlere ve time gre total_bill deerlerinin toplamn, min, max ve ortalamasn bulunuz.
#########################################
time_day_group = tips.groupby(["day","time"])["total_bill"].agg(["sum","min","max","mean"])
print(time_day_group)

#########################################
# Grev 20:Lunch zamanna ve kadn mterilere ait total_bill ve tip  deerlerinin day'e gre toplamn, min, max ve ortalamasn bulunuz.
#########################################
lunch_females = tips[(tips['time'] == 'Lunch') & (tips['sex'] == 'Female')]
lunch_female_stats = lunch_females.groupby('day')[['total_bill', 'tip']].agg(['sum', 'min', 'max', 'mean'])
print(lunch_female_stats)
#########################################
# Grev 21: size'i 3'ten kk, total_bill'i 10'dan byk olan siparilerin ortalamas nedir?
#########################################

filtered_orders = tips[(tips['size'] < 3) & (tips['total_bill'] > 10)]
average_total_bill = filtered_orders['total_bill'].mean()
print(average_total_bill)

#########################################
# Grev 22: total_bill_tip_sum adnda yeni bir deiken oluturun. Her bir mterinin dedii totalbill ve tip in toplamn versin.
#########################################

tips['total_bill_tip_sum'] = tips['total_bill'] + tips['tip']
print(tips[['total_bill', 'tip', 'total_bill_tip_sum']].head())

#########################################
# Grev 23: total_bill_tip_sum deikenine gre bykten ke sralaynz ve ilk 30 kiiyi yeni bir dataframe'e ataynz.
#########################################



```
